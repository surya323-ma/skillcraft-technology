Retail Customer Segmentation with K-means

```python
!pip install kagglehub scikit-learn --quiet
import kagglehub
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
```

Step 1: Download Dataset from Kaggle

# Authenticate and download dataset

```python
print("Downloading dataset...")
path = kagglehub.dataset_download("vjchoudhary7/customer-segmentation-tutorial-in-python")
print(f"Dataset downloaded to: {path}")
```

Step 2: Load and Explore Data

```python
df = pd.read_csv(f"{path}/Mall_Customers.csv")
print("First 5 rows:")
df.head()
```

```python
print("\nData Summary:")
print(f"Shape: {df.shape}")
print("\nMissing values:")
print(df.isna().sum())
print("\nData Types:")
print(df.dtypes)
```

```python
plt.figure(figsize=(15, 4))
plt.subplot(1, 3, 1)
sns.histplot(df['Age'], bins=20, kde=True)
plt.title('Age Distribution')
plt.subplot(1, 3, 2)
sns.histplot(df['Annual Income (k$)'], bins=20, kde=True)
plt.title('Income Distribution')
plt.subplot(1, 3, 3)
sns.histplot(df['Spending Score (1-100)'], bins=20, kde=True)
plt.title('Spending Score Distribution')
plt.tight_layout()
plt.show()
```

step 3 :data prerocessing

```python
X = df[['Annual Income (k$)', 'Spending Score (1-100)']].values
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
print("First 5 scaled samples:")
print(X_scaled[:5])
```

Step 4: Determine Optimal Clusters

```python
wcss = []
sil_scores = []
for i in range(2, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)
    sil_scores.append(silhouette_score(X_scaled, kmeans.labels_))
```

```python
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(range(2, 11), wcss, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.subplot(1, 2, 2)
plt.plot(range(2, 11), sil_scores, marker='o')
plt.title('Silhouette Scores')
plt.xlabel('Number of clusters')
plt.ylabel('Score')
plt.tight_layout()
plt.show()
```

Step 5: Train K-means Model (5 clusters)

```python
kmeans = KMeans(n_clusters=5, random_state=42)
df['Cluster'] = kmeans.fit_predict(X_scaled)
```

```python
centroids = scaler.inverse_transform(kmeans.cluster_centers_)
```

Step 6: Visualize Results

```python
colors = ['#FF5733', '#33FF57', '#3357FF', '#F333FF', '#33FFF3']
cluster_names = {
    0: 'Budget Shoppers',
    1: 'Standard Spenders',
    2: 'Target Customers',
    3: 'Carefree Spenders',
    4: 'Conservative Shoppers'
}
```

```python
plt.figure(figsize=(10, 6))
for i in range(5):
    cluster_data = df[df['Cluster'] == i]
    plt.scatter(
        cluster_data['Annual Income (k$)'],
        cluster_data['Spending Score (1-100)'],
        s=100, c=colors[i],
        label=cluster_names[i]
    )
```

```python
plt.scatter(
    centroids[:, 0], centroids[:, 1],
    s=200, marker='X', c='black',
    label='Centroids'
)
plt.title('Customer Segmentation', fontsize=14)
plt.xlabel('Annual Income (k$)', fontsize=12)
plt.ylabel('Spending Score (1-100)', fontsize=12)
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

Step 7: Cluster Analysis

```python
cluster_stats = df.groupby('Cluster').agg({
    'Age': 'median',
    'Annual Income (k$)': ['mean', 'std'],
    'Spending Score (1-100)': ['mean', 'std'],
    'Gender': lambda x: x.mode()[0]
}).round(1)
cluster_stats['Count'] = df['Cluster'].value_counts()
cluster_stats['Segment'] = [cluster_names[i] for i in range(5)]
print("Cluster Characteristics:")
cluster_stats
```
